```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df=sns.load_dataset('penguins')
```

```python
df.head()
```

```python
df.tail()
```

## Here we have to check whether the data set contains null values if it has null values we may get error so we have to remove the null values or we can fill it with avg value 

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.columns
```

```python
df.head()
```

```python
print(df['island'].unique())
```

```python
pd.get_dummies(df['sex']).head()
```

```python
df.head()
```

```python
df['sex']=df['sex'].str.strip().str.capitalize().map({"Male":1,"Female":0})
```

```python
df.head()
```

```python
df['species'].unique()
```

```python
df['species']=df['species'].map({'Adelie':0,'Chinstrap':1,'Gentoo':2})
```

```python
df['island'].unique()
```

```python
df['island']=df['island'].map({'Torgersen':0,'Biscoe':1,'Dream':2})
```

```python
df.head()
```

```python
x=df.drop(columns=['species'])
```

```python
y=df['species']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_test.shape)
print(y_train.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=100,random_state=42,criterion='entropy')
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred[:10]
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
acc=accuracy_score(y_test,y_pred)
```

```python
print(acc*100)
```

```python
classifi=classification_report(y_test,y_pred)
```

```python
print(classifi)
```
